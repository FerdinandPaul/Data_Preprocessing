The 3 plots below are based on data from the Twitter archive of the site WeRateDogs and data added by an API from the same source. Although it is nice to see what names people give their dogs (Plot1), it is more interesting to see at what hour most tweets are posted (Plot2). From this it can be concluded that most pictures are uploaded when people are working. You can see from the timeline (Plot 2) that most posts are made in the early afternoon and at midnight. So we can conclude that people who tweet on WeRateDogs of course have a dog, mostly work and probably have family, because in the evening, where the family sits together and eats, fewer posts are made. The last plot (Plot3) shows that there is a correlation between the most favorited tweets and the most retweeted ones. The more often a tweet is favorited by users, the more often the same post is retweeted. From this we can conclude that certain staging, as well as certain dog breeds or popular users benefit. As a further step, we could find out how much the dog breed is responsible for the popularity of the tweets. If we take a closer look at the names (Plot 2), we can see that certain names like Cahrlie, Lucy or Cooper are very popular. However, despite the many rows in the DataFrame, we can see that the most popular name occurs only 11 times and other names repeat even less.
